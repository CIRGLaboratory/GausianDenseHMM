```python
import numpy as np
import pandas as pd
import pickle
import os
```

```python
os.listdir("/ziob/kabalce/GausianDenseHMM/data/")
```

```python
result_path = "/ziob/kabalce/GausianDenseHMM/data/benchmark_artificial-2022-11-10-iiii"
result_files = os.listdir(result_path)

results = dict()
for file_name in result_files:
    with open(f"{result_path}/{file_name}", "rb") as f:
        results[file_name.split(".")[0]] = pickle.load(f)
results
```

```python
summary = []
for k, v in results.items():
    summary.append({kk: v[kk] for kk in ['n', 'd', 'T', 'loglikelihood', 'omega_loss', 'accuracy', 'time']})
summary = pd.DataFrame(summary, columns=['n', 'd', 'T', 'loglikelihood', 'omega_loss', 'accuracy', 'time'])
```
